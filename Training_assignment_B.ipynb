考察  
 売上金額と企業等数・事業所数・従業者数の相関係数はいずれも0.9を超えており、強い正の相関が見られる。
 
 
 売上金額  
 千代田区・港区・中央区など、山手線の内側および沿線に位置する23区の中心部がトップを占めている。  
 その下には北区・足立区・荒川区など、23区の外縁部の区が名を連ねている。  
 その次には八王子市・府中市・立川市など、23区から距離がある多摩地域の拠点都市が続いている。  
 さらにその下には、国分寺市・東大和市・清瀬市など、23区と多摩地域の間にあり、両地域へ通勤する人々のベッドタウンである地域がある。  
 最後に、八丈島・三宅島・小笠原諸島などの離島部が名を連ねる。  
 
 企業等数  
 大田区が1位となっている。大田区は主に製造業の下請けを担う中小企業が集中しているため、企業の数でトップになったと思われる。同様の理由で足立区なども売上金額に比べて上位に食い込んでいる。  
 また、八王子市・町田市・府中市などの多摩地域の拠点都市が比較的上位にいる。23区ではなく、その周辺部に本社を置く事で、地域に根ざした事業を行う企業が一定数あることがうかがえる。23区の中心部に比べて建設業や製造業が多いことからも裏付けられる。  
 
 事業所数  
 概ね売上金額と同様の分布となっている。しかし、わずかながら売上金額の順位との相違も見られる。世田谷区・足立区・八王子市などは順位を上げたのに対し、江東区・目黒区・中野区などは順位を下げている。  
 順位を上げた市区町村の特徴として、建設業と宿泊業が比較的大きな割合を占めていることがあげられる。東京オリンピックによる建設需要の増加やインバウンド需要が関係していると思われる。  
 反対に順位を下げたところは主に卸売業がメインである場合が多い。長引く不景気で出荷数が伸び悩んでいるのではないだろうか。  
 
 従業員数  
 こちらも概ね売上金額と同様の分布となっている。だが、八王子・町田・小平などの順位が相対的に高く、売上の分布よりも広がりが大きいように見受けられる。要因の一つとして、23区中心部の業種はサービス業などのホワイトカラーが多く従業員一人あたりの売上が大きくなる一方、郊外の業種は建設や製造などブルーカラーの割合が相対的に大きくなり、従業員一人当たり売上がそこまで伸びないと考えられる。そのため、相対的に郊外の従業員数は多くなると考えられる。
